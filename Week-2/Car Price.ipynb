## Calculating correlation matrix

```python
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        text = f"{corr_matrix.iloc[i, j]:.2f}"
        plt.text(j, i, text, ha='center', va='center', color='black')

plt.title('Correlation Matrix with Values')
plt.show()
```

## Distribution of Car Prices

```python
plt.figure(figsize=(12, 6))
plt.hist(df['price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices')
plt.show()
```

## Count of each cars

```python
plt.figure(figsize=(12, 6))
car_counts[car_counts>1].plot(kind='bar')
plt.xlabel('Car Name')
plt.ylabel('Count')
plt.title('Count of Cars')
plt.xticks(rotation=90)
plt.show()
```

## Categorical variables analysis

### Calculating frequency of fueltype, aspiration, carbody and drivewheel

```python
categorical_cols = ['fueltype', 'aspiration', 'carbody', 'drivewheel']
for col in categorical_cols:
   
    category_counts = df[col].value_counts()
    
    plt.figure(figsize=(8, 6))
    category_counts.plot(kind='bar')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.title('Frequency of Categories: ' + col)
    plt.show()
```

## Distribution of car prices for each Car body type

```python
avg_price_by_body = df.groupby('carbody')['price'].mean().reset_index()
avg_price_by_body = avg_price_by_body.sort_values('price')
plt.figure(figsize=(8, 6))
plt.bar(avg_price_by_body['carbody'], avg_price_by_body['price'])
plt.xlabel('Car Body Type')
plt.ylabel('Average Price')
plt.title('Average Car Price by Body Type')
plt.xticks(rotation=45)
plt.show()
```

## Analyze the correlation between car price and mileage

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['citympg'], df['price'])
plt.xlabel('City MPG')
plt.ylabel('Price')
plt.title('Car Price vs City MPG')
plt.show()
```

## Relationship between the car price and the horsepower of the car

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['horsepower'], df['price'])
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Car Price vs Horsepower')
plt.show()
```

## Relationships between car price and specific features like the compression ratio, boreratio, or stroke

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.scatter(df['compressionratio'], df['price'])
plt.xlabel('Compression Ratio')
plt.ylabel('Price')
plt.title('Car Price vs Compression Ratio')

plt.subplot(1, 3, 2)
plt.scatter(df['boreratio'], df['price'])
plt.xlabel('Bore Ratio')
plt.ylabel('Price')
plt.title('Car Price vs Bore Ratio')

plt.subplot(1, 3, 3)
plt.scatter(df['stroke'], df['price'])
plt.xlabel('Stroke')
plt.ylabel('Price')
plt.title('Car Price vs Stroke')

plt.tight_layout()
plt.show()
```
